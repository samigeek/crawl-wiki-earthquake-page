```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/List_of_deadly_earthquakes_since_1900'
response = requests.get(url)

print(response.encoding)

```

```python
soup = BeautifulSoup(response.text, 'html.parser')
table=soup.find('table',class_='sortable wikitable')
th = table.findAll('th')
col=[i.get_text() for i in th]
col=[i[:-1] for i in col]

rows = table.findAll('tr')
r=[]

for row in rows:
    body=row.findAll('td')
    r.append([i.get_text() for i in body])
r.pop(0);
```

```python
df= pd.DataFrame(r,columns=col)
df.head()
```

```python
#non values
df1 = df1.replace(['',float('nan')],np.nan)
#drop nan values from laybel column
df1=df1.dropna(subset=['death'])
#change df type to int type
df1['death'] = df1['death'].apply(lambda x: int(np.ceil(x)))
# '' in Secondary Effects column means the earthquake hasnt any Secondary Effects so we fill it with 'nothing'
df1['Secondary Effects']=df1['Secondary Effects'].fillna('nothing')

df1['Secondary Effects']=df1['Secondary Effects'].map({'F': 'fire', 'L': 'landslide','LF':'l,f','LFLq':'l,f,lq','LLq':'l,lq','Lq':'liquefaction','LqL':'l,lq','T':'tsunami',
                              'TF':'t,f','TL':'t,l','TLF':'t,l,f','TLLq':'t,l,lq','TLq':'t,lq','nothing':'nothing'})
df1=df1.rename(columns={'Present-day country and link to Wikipedia article':'country'})
df1.head()
```

```python
def crawl(df0,x):
    link=df0['country'].iloc[x].split()
    if '(see' not in link:
        return False
    link=link[link.index('(see')+1:]
    url = 'https://en.wikipedia.org/wiki/'+"_".join(link)[:-1]
    response1 = requests.get(url)
    soup1 = BeautifulSoup(response1.text, 'html.parser')
    return soup1
```

```python
#find lat & long from crawling country links
df2=df1.copy()
df2.reset_index(drop=True,inplace=True)
geo_fill=df2.index[(df2['Lat'].isna())&(df2['death'].notna())]

for i in geo_fill:
    so=crawl(df2,i)
    class_=so.find(class_='geo-default')
    geo_att=class_.find(title='Maps, aerial photos, and other data for this location').get_text().split()
    df2['Lat'].iloc[i]=geo_att[0][:5]
    df2['Long'].iloc[i]=geo_att[1][:5]
    df2['Depth (km)'].iloc[i]=so.find_all(class_='infobox-data')[6].get_text()[:5]
```

```python
#finding local time
from dateutil import tz
from timezonefinder import TimezoneFinder
def local(x):

    #find time zone by lat and long
    tf = TimezoneFinder(in_memory=True)
    local_time_zone = tf.timezone_at(lng=x['Long'], lat=x['Lat'])

    #finding local time by time zone
    from_zone = tz.gettz('UTC')
    to_zone = tz.gettz(local_time_zone)
    utc=x['Origin (UTC)']
    utc = utc.replace(tzinfo=from_zone)

    # Convert time zone
    central = utc.astimezone(to_zone)
    return central


df4=df3.copy()
df4['local time']=df4.apply(lambda x :local(x), axis=1)
df4.head()
```
